## **Getting started**
</div>

The document you are reading is not a static web page, but an interactive environment called a **Colab notebook** that lets you write and execute code.

For example, here is a **code cell** with a short Python script that computes a value, stores it in a variable, and prints the result:

```
seconds_in_a_day = 24 * 60 * 60
seconds_in_a_day
```

To execute the code in the above cell, select it with a click and then either press the play button to the left of the code, or use the keyboard shortcut "Command/Ctrl+Enter". To edit the code, just click the cell and start editing.

Variables that you define in one cell can later be used in other cells:

```
seconds_in_a_week = 7 * seconds_in_a_day
seconds_in_a_week
```

Colab notebooks allow you to combine **executable code** and **rich text** in a single document, along with **images**, **HTML**, **LaTeX** and more. When you create your own Colab notebooks, they are stored in your Google Drive account. You can easily share your Colab notebooks with co-workers or friends, allowing them to comment on your notebooks or even edit them. To learn more, see [Overview of Colab](/notebooks/basic_features_overview.ipynb). ## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```
# Group the data by country code and year, and sum the values
country_data = df.groupby(['code', 'year']).sum().reset_index()

# Calculate the total tourism receipts for each country over the entire period
country_totals = country_data.groupby('code').sum().reset_index()

# Sort the countries by their total tourism receipts in descending order
sorted_countries = country_totals.sort_values('value_$', ascending=False)

# Select the top 20 countries with the highest tourism receipts
top_20_countries = sorted_countries.head(20)

# Select the Asian countries from the top 20 list
asian_countries = top_20_countries[top_20_countries['code'].isin(['CHN', 'HKG', 'IND', 'JPN', 'KOR', 'MYS', 'PHL', 'SGP', 'THA', 'TWN'])]

# Filter the data to include only the selected countries
selected_data = df[df['code'].isin(top_20_countries['code'])]

# Further filter the data to include only the selected Asian countries
selected_asian_data = df[df['code'].isin(asian_countries['code'])]
```

```
# Select the top 20 countries by tourism receipts
top20 = df.groupby('name')['value_$'].sum().sort_values(ascending=False).head(20)
top20.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Tourism Receipts ($)')
plt.title('Total Tourism Receipts for Top 20 Countries')
plt.show()
```

```
# Select the top 10 countries by tourism receipts
top10 = df.groupby('name')['value_$'].sum().sort_values(ascending=False).head(10)

# Create a line plot of tourism receipts by year for each of the top 10 countries
df_top10 = df[df['name'].isin(top10.index)]
fig, ax = plt.subplots(figsize=(10, 6))
for country in top10.index:
    ax.plot(df_top10[df_top10['name'] == country]['year'], df_top10[df_top10['name'] == country]['value_$'], label=country)
ax.set_xlabel('Year')
ax.set_ylabel('Tourism Receipts ($)')
ax.set_title('Tourism Receipts by Year for Top 10 Countries')
ax.legend()
plt.show()

# Create a stacked bar chart of tourism receipts by year for each of the top 10 countries
df_top10_pivot = df_top10.pivot(index='year', columns='name', values='value_$')
fig, ax = plt.subplots(figsize=(10, 6))
df_top10_pivot.plot(kind='bar', stacked=True, ax=ax)
ax.set_xlabel('Year')
ax.set_ylabel('Tourism Receipts ($)')
ax.set_title('Tourism Receipts by Year for Top 10 Countries')
plt.show()

# Create a scatter plot of tourism receipts by year for each of the top 10 countries
fig, ax = plt.subplots(figsize=(10, 6))
for country in top10.index:
    ax.scatter(df_top10[df_top10['name'] == country]['year'], df_top10[df_top10['name'] == country]['value_$'], label=country)
ax.set_xlabel('Year')
ax.set_ylabel('Tourism Receipts ($)')
ax.set_title('Tourism Receipts by Year for Top 10 Countries')
ax.legend()
plt.show()
```

```
# Select the countries of interest
countries = ['Pakistan', 'India', 'Iran', 'Afghanistan', 'Sri Lanka', 'China', 'Bangladesh']
df_selected = df[df['name'].isin(countries)]

# Create a line plot of tourism receipts by year for each country
sns.lineplot(x='year', y='value_$', hue='name', data=df_selected)
plt.xlabel('Year')
plt.ylabel('Tourism Receipts ($)')
plt.title('Tourism Receipts by Year for Selected Countries')
plt.show()
```

```
# Select the countries of interest
countries = ['Pakistan', 'India', 'Iran', 'Afghanistan', 'Sri Lanka', 'China', 'Bangladesh']
df_countries = df[df['name'].isin(countries)]
# Create a stacked bar chart of tourism receipts by year for each country
df_pivot = df_countries.pivot(index='year', columns='name', values='value_$')
fig, ax = plt.subplots(figsize=(10, 6))
df_pivot.plot(kind='bar', stacked=True, ax=ax)
ax.set_xlabel('Year')
ax.set_ylabel('Tourism Receipts ($)')
ax.set_title('Tourism Receipts by Year for Selected Countries')
plt.show()
```
